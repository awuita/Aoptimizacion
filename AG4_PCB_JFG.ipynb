## Función objetivo

Dado que el objetivo es optimizar los ingresos por visualizaciones, nos enfrentamos a un problema de **maximización**.

Para evaluar la calidad de cada solución propuesta por nuestro algoritmo, hemos diseñado una función objetivo que cuantifica la audiencia total de la jornada (en millones de espectadores). Esta función suma la audiencia de cada partido tras aplicar los factores de corrección por horario y por solapamiento:

Para cada partido `i` con franja asignada `h_i`, su audiencia real es:

$$\text{audiencia}_i = A_i \times c_{h_i} \times p_{n_{h_i}}$$

donde:
- $A_i$ es la audiencia base según las categorías de los equipos (en millones)
- $c_{h_i}$ es el coeficiente del horario asignado (va de 0.40 a 1.00)
- $p_{n_{h_i}}$ es la penalización por coincidir con otros partidos en la misma franja

La función objetivo suma las audiencias de los 10 partidos:

$$FO(\mathbf{h}) = \sum_{i=0}^{9} A_i \times c_{h_i} \times p_{n_{h_i}}$$

Para que una solución sea válida debe cumplir:
- Al menos un partido en la franja 0 (Viernes 20h)
- Al menos un partido en la franja 9 (Lunes 20h)

El objetivo es encontrar la asignación $\mathbf{h}^*$ que maximice $FO$ entre todas las soluciones válidas.


Diseña un algoritmo para resolver el problema por fuerza bruta

````python
from itertools import product
from collections import Counter

# 1. Parámetros globales y estructuras auxiliares
# Definimos las franjas horarias con sus respectivos coeficientes de ponderación
franjas = [
    ("Viernes 20h",  0.40),  # 0 <- obligatorio
    ("Sabado 12h",   0.55),  # 1
    ("Sabado 16h",   0.70),  # 2
    ("Sabado 18h",   0.80),  # 3
    ("Sabado 20h",   1.00),  # 4 <- referencia (mejor horario)
    ("Domingo 12h",  0.45),  # 5
    ("Domingo 16h",  0.75),  # 6
    ("Domingo 18h",  0.85),  # 7
    ("Domingo 20h",  1.00),  # 8
    ("Lunes 20h",    0.40),  # 9 <- obligatorio
]

# Audiencias base por tipo de enfrentamiento.
# Hemos incluido ambas permutaciones (A,B y B,A) para agilizar la consulta
# y evitar comprobaciones adicionales dentro del bucle crítico de la función objetivo.
aud_base = {
    ("A","A"): 2.0,
    ("A","B"): 1.3, ("B","A"): 1.3,
    ("A","C"): 1.0, ("C","A"): 1.0,
    ("B","B"): 0.75,
    ("B","C"): 0.47, ("C","B"): 0.47,
    ("C","C"): 0.47,
}

# Coeficiente de reducción según cuántos partidos coinciden en la misma franja.
# Usamos un diccionario para aplicar directamente el porcentaje de reducción
# sobre la audiencia base según el número de coincidencias.
pen_coinc = {0: 1.00, 1: 0.75, 2: 0.55, 3: 0.40,
             4: 0.30, 5: 0.25, 6: 0.22, 7: 0.20, 8: 0.20}

def penalizacion(n):
    # Calcula la reducción de audiencia restando 1 al total de partidos
    # en la franja para obtener el número de coincidencias reales.
    return pen_coinc.get(n - 1, 0.20)

def fo(asig, partidos):
    # Función objetivo que calcula la audiencia total de una jornada.
    # Usamos Counter para saber cuántos partidos hay por horario y aplicar las penalizaciones.
    cnt = Counter(asig)
    total = 0.0
    for i, h in enumerate(asig):
        total += aud_base[partidos[i]] * franjas[h][1] * penalizacion(cnt[h])
    return round(total, 4)

def valida(asig):
    # Comprobamos que la solución cumple las restricciones del enunciado:
    # debe haber al menos un partido el viernes (0) y otro el lunes (9).
    return 0 in asig and 9 in asig

def fuerza_bruta(partidos, n_franjas=10):
    # Algoritmo de fuerza bruta que explora todas las combinaciones posibles.
    best, best_fo, n_eval = None, -1, 0
    for asig in product(range(n_franjas), repeat=len(partidos)):
        if valida(asig):
            v = fo(list(asig), partidos)
            n_eval += 1
            if v > best_fo:
                best_fo, best = v, list(asig)
    return best, best_fo, n_eval

# 2. Ejecución y Pruebas
# Pruebas con una sub-instancia pequeña
# Hemos reducido la jornada a 4 partidos para que la fuerza bruta sea viable
demo = [("A","B"), ("B","B"), ("B","C"), ("C","C")]

sol, res, n_eval = fuerza_bruta(demo)

print("Combinaciones válidas evaluadas:", n_eval)
print("El máximo de audiencia conseguido es:", res, "millones")
print("La mejor asignación de horarios es:", sol)

print("\nDetalle de la jornada:")
for i in range(len(sol)):
    horario_idx = sol[i]
    nombre_horario = franjas[horario_idx][0]
    partido = demo[i]
    print("Partido", i, partido, "se juega el:", nombre_horario)
````

## Complejidad del algoritmo de fuerza bruta

Con $N = 10$ partidos y $H = 10$ franjas, el total de combinaciones posibles es:

$$H^N = 10^{10} = 10.000.000.000$$

Para cada una hay que verificar las restricciones y calcular la función objetivo, ambas operaciones en $O(N)$. La complejidad total es por tanto:

$$O(H^N \cdot N)$$

Con $H = N = 10$ eso son del orden de $10^{11}$ operaciones elementales.

Si consideramos que en un entorno de ejecución estándar con Python podemos procesar aproximadamente 106 evaluaciones por segundo, el tiempo estimado para explorar todo el espacio sería de unas 2.7 horas.

Por este motivo, en la prueba anterior limitamos la jornada a 4 partidos (104 combinaciones), para poder verificar que el código es correcto sin esperar horas.

Esto confirma que la fuerza bruta no es una opción práctica para el problema completo y que hay que recurrir a algún método heurístico.

## Por qué la Búsqueda Tabú mejora la fuerza bruta

La fuerza bruta explora todas las combinaciones posibles del espacio de soluciones, lo que con 10 partidos y 10 franjas supone $10^{10}$ estados. Es computacionalmente inviable para la instancia completa del problema.

La Búsqueda Tabú en cambio no explora todo el espacio: parte de una solución aleatoria válida y en cada iteración solo evalúa los 90 vecinos inmediatos (cambiar la franja de un partido). Elige el mejor de esos vecinos y se mueve a él, aunque no mejore la solución actual. La clave es la **lista tabú**: un registro de los últimos movimientos realizados que están temporalmente prohibidos. Esto evita que el algoritmo dé vueltas en círculo y le obliga a explorar zonas nuevas del espacio que de otro modo no visitaría.

Además, incorporamos el **criterio de aspiración**: si un movimiento tabú conduciría a la mejor solución encontrada hasta ahora, se acepta igualmente. Esto garantiza que nunca descartamos una mejora real por culpa de la restricción tabú.

En resumen, en lugar de explorar los $10^{10}$ estados, exploramos como máximo $500 \times 90 = 45.000$, lo que supone una reducción drástica manteniendo una calidad de solución razonablemente buena.

````python
import random
from collections import deque

# Búsqueda Tabú
# La idea es parecida a la búsqueda local clásica, pero con una diferencia importante:
# llevamos un registro de los últimos movimientos realizados (lista tabú) para no repetirlos.
# Esto obliga al algoritmo a explorar zonas nuevas del espacio aunque no haya mejora inmediata,
# lo que le permite escapar de los óptimos locales en los que se quedaría atascado una búsqueda sin memoria.

def asig_aleatoria(n=10, h=10):
    # Generamos una asignación al azar y validamos que cumpla las restricciones de la jornada
    while True:
        asig = [random.randint(0, h - 1) for _ in range(n)]
        if valida(asig):
            return asig

def vecinos(asig, h=10):
    # Definimos el vecindario mediante el cambio de la franja horaria de un único partido.
    # Junto a cada vecino guardamos el movimiento que lo generó: (partido, franja_nueva)
    # para poder añadirlo a la lista tabú si lo elegimos.
    res = []
    for i in range(len(asig)):
        for franja in range(h):
            if franja != asig[i]:
                v = asig[:]
                v[i] = franja
                res.append((v, (i, franja)))
    return res

def busqueda_tabu(partidos, h=10, max_iter=500, tamanio_tabu=15):
    # Iniciamos desde una solución aleatoria válida
    asig = asig_aleatoria(len(partidos), h)
    actual_fo = fo(asig, partidos)

    # Guardamos siempre la mejor solución encontrada, aunque no sea la actual
    mejor_asig = asig[:]
    mejor_fo   = actual_fo

    # La lista tabú almacena los últimos movimientos prohibidos.
    # Usamos deque con tamaño fijo para que los movimientos más antiguos
    # se expulsen solos cuando la lista se llena.
    lista_tabu = deque(maxlen=tamanio_tabu)

    for _ in range(max_iter):
        candidatos = vecinos(asig, h)

        # Elegimos el mejor vecino válido que no esté en la lista tabú.
        # Si ninguno cumple, nos quedamos con el mejor aunque esté prohibido
        # (criterio de aspiración: si mejora el mejor global, se acepta igualmente).
        mejor_vecino     = None
        mejor_fo_vecino  = -1
        mov_elegido      = None

        for v, mov in candidatos:
            if not valida(v):
                continue
            v_fo = fo(v, partidos)
            # Aceptamos el movimiento si no está en tabú O si mejora el mejor global (aspiración)
            if mov not in lista_tabu or v_fo > mejor_fo:
                if v_fo > mejor_fo_vecino:
                    mejor_vecino    = v
                    mejor_fo_vecino = v_fo
                    mov_elegido     = mov

        # Si no encontramos ningún candidato viable, terminamos
        if mejor_vecino is None:
            break

        # Nos movemos al mejor vecino aunque sea peor que el actual
        asig       = mejor_vecino
        actual_fo  = mejor_fo_vecino
        lista_tabu.append(mov_elegido)

        # Actualizamos el mejor global si hemos mejorado
        if actual_fo > mejor_fo:
            mejor_fo   = actual_fo
            mejor_asig = asig[:]

    return mejor_asig, mejor_fo

# Ejecución final del algoritmo con los parámetros configurados para la jornada completa
print("Ejecutando Búsqueda Tabú...")
jornada_completa = [("A","B"), ("A","C"), ("B","B"), ("B","C"), ("B","C"),
                    ("B","B"), ("C","C"), ("B","C"), ("A","B"), ("A","A")]

sol, aud = busqueda_tabu(jornada_completa)

print("Resultado Búsqueda Tabú:")
print(f"Audiencia total: {aud} millones")
print("Asignación final:")

for i in range(len(sol)):
    partido = jornada_completa[i]
    horario = franjas[sol[i]][0]
    print(f"Partido {i+1} ({partido[0]}-{partido[1]}): {horario}")
````

## Complejidad del algoritmo mejorado

Para mejorar la fuerza bruta se utiliza una **Búsqueda Tabú**, que es una técnica de búsqueda local con memoria. La idea principal es explorar el espacio de soluciones evitando volver a estados recientes y así no quedarse atrapado en óptimos locales.

En cada iteración se evalúa todo el vecindario de la solución actual. En este problema, el vecindario tiene $N \times (H-1) = 10 \times 9 = 90$ vecinos, ya que se cambia la franja horaria de un único partido cada vez. De todos ellos se selecciona el mejor movimiento que no esté prohibido por la lista tabú, aunque no mejore la solución actual.

Si el algoritmo se ejecuta durante un máximo de $I$ iteraciones, su complejidad es:

$$
O(I \times N \times H)
$$

Con los valores usados ($I = 500$, $N = 10$, $H = 10$), se realizan como máximo 50.000 evaluaciones de la función objetivo, muy lejos de las $10^{10}$ combinaciones de la fuerza bruta.

La lista tabú evita ciclos y permite diversificar la búsqueda sin necesidad de reinicios aleatorios. En este caso, un tamaño de lista tabú de 15 movimientos da buenos resultados. Aunque no garantiza el óptimo global, la calidad de las soluciones obtenidas es buena con un coste computacional asumible.

Según el problema (y tenga sentido), diseña un juego de datos de entrada aleatorios

````python
import random
from collections import Counter

# Definimos las franjas horarias con sus respectivos pesos de audiencia
franjas = [
    ("Viernes 20h",  0.40),
    ("Sabado 12h",   0.55),
    ("Sabado 16h",   0.70),
    ("Sabado 18h",   0.80),
    ("Sabado 20h",   1.00),
    ("Domingo 12h",  0.45),
    ("Domingo 16h",  0.75),
    ("Domingo 18h",  0.85),
    ("Domingo 20h",  1.00),
    ("Lunes 20h",    0.40),
]

# Estructura para consultar la audiencia base según el cruce de categorías
aud_base = {
    ("A","A"): 2.0,
    ("A","B"): 1.3, ("B","A"): 1.3,
    ("A","C"): 1.0, ("C","A"): 1.0,
    ("B","B"): 0.75,
    ("B","C"): 0.47, ("C","B"): 0.47,
    ("C","C"): 0.47,
}

# Coeficientes de penalización por solapamiento de partidos en la misma franja
pen_coinc = {0: 1.00, 1: 0.75, 2: 0.55, 3: 0.40,
             4: 0.30, 5: 0.25, 6: 0.22, 7: 0.20, 8: 0.20}

def penalizacion(n):
    # Restamos 1 al total de partidos para saber cuántos coinciden realmente y aplicamos la reducción.
    return pen_coinc.get(n - 1, 0.20)

def fo(asig, partidos):
    # Calculamos la audiencia total de la jornada aplicando todos los coeficientes
    cnt = Counter(asig)
    total = 0.0
    for i, h in enumerate(asig):
        total += aud_base[partidos[i][2]] * franjas[h][1] * penalizacion(cnt[h])
    return round(total, 4)

def valida(asig):
    # Comprobamos que se cumpla la restricción obligatoria de partidos en Viernes y Lunes
    return 0 in asig and 9 in asig

# Listado de equipos de LaLiga con su clasificación por categoría de audiencia
equipos = [
    ("Real Madrid", "A"), ("Barcelona", "A"), ("Atlético", "A"),
    ("Sevilla", "B"), ("Alavés", "B"), ("Athletic", "B"),
    ("Villarreal", "B"), ("Real Sociedad", "B"), ("Valencia", "B"),
    ("Getafe", "B"), ("Espanyol", "B"), ("Betis", "B"),
    ("Celta", "B"), ("Levante", "B"),
    ("Leganés", "C"), ("Valladolid", "C"), ("Eibar", "C"),
    ("Granada", "C"), ("Mallorca", "C"), ("Osasuna", "C")
]

# Mostramos la distribución inicial de equipos
print("Equipos por categoria:")
for cat in ["A", "B", "C"]:
    grupo = [e[0] for e in equipos if e[1] == cat]
    print(f"Cat {cat}: {', '.join(grupo)}")

def generar_jornada(equipos, seed=None):
    # Mezclamos los equipos para generar una jornada aleatoria
    if seed is not None:
        random.seed(seed)
    lista = equipos[:]
    random.shuffle(lista)
    jornada = []
    # Emparejamos los equipos de dos en dos para formar los 10 partidos
    for i in range(0, 20, 2):
        loc, cat_l = lista[i]
        vis, cat_v = lista[i + 1]
        tipo = tuple(sorted([cat_l, cat_v]))
        jornada.append((loc, vis, tipo))
    return jornada

# Generamos la instancia de la jornada que utilizaremos en las pruebas
jornada = generar_jornada(equipos, seed=42)

print("\nPartidos de la jornada generada:")
for i, (loc, vis, tipo) in enumerate(jornada):
    base = aud_base[tipo]
    print(f"{i+1}. {loc} vs {vis} ({tipo[0]}-{tipo[1]}) -> {base}M")

# Calculamos la audiencia base teórica sin penalizaciones
total_base = sum(aud_base[p[2]] for p in jornada)
print(f"\nAudiencia base total (sin solapamientos): {total_base:.2f}M")

# Resumen de los tipos de enfrentamientos
tipos = Counter(p[2] for p in jornada)
print("Frecuencia de enfrentamientos:", dict(tipos))
````

Aplica el algoritmo al juego de datos generado

````python
import time

# Extraemos solo las categorías de la jornada para facilitar el proceso de optimización
cats = [p[2] for p in jornada]

def asig_aleatoria(n=10, h=10):
    # Generamos soluciones aleatorias hasta que una cumpla las restricciones de la liga
    while True:
        asig = [random.randint(0, h - 1) for _ in range(n)]
        if valida(asig):
            return asig

def vecinos(asig, h=10):
    # Construimos el vecindario modificando la franja de un partido cada vez.
    # Devolvemos también el movimiento realizado para poder registrarlo en la lista tabú.
    res = []
    for i in range(len(asig)):
        for franja in range(h):
            if franja != asig[i]:
                v = asig[:]
                v[i] = franja
                res.append((v, (i, franja)))
    return res

def busqueda_tabu(partidos, h=10, max_iter=500, tamanio_tabu=15):
    # Iniciamos desde un punto aleatorio válido
    asig = asig_aleatoria(len(partidos), h)
    actual_fo = fo(asig, [(None, None, c) for c in partidos])

    mejor_asig = asig[:]
    mejor_fo   = actual_fo

    # La lista tabú bloquea los últimos movimientos para forzar la exploración de zonas nuevas
    lista_tabu = deque(maxlen=tamanio_tabu)

    for _ in range(max_iter):
        candidatos = vecinos(asig, h)

        mejor_vecino    = None
        mejor_fo_vecino = -1
        mov_elegido     = None

        for v, mov in candidatos:
            if not valida(v):
                continue
            v_fo = fo(v, [(None, None, c) for c in partidos])
            # Aceptamos si no está en tabú, o si mejora el mejor global (criterio de aspiración)
            if mov not in lista_tabu or v_fo > mejor_fo:
                if v_fo > mejor_fo_vecino:
                    mejor_vecino    = v
                    mejor_fo_vecino = v_fo
                    mov_elegido     = mov

        if mejor_vecino is None:
            break

        # Nos movemos al mejor candidato aunque empeore respecto al estado actual
        asig      = mejor_vecino
        actual_fo = mejor_fo_vecino
        lista_tabu.append(mov_elegido)

        # Guardamos la mejor solución global encontrada hasta ahora
        if actual_fo > mejor_fo:
            mejor_fo   = actual_fo
            mejor_asig = asig[:]

    return mejor_asig, mejor_fo

# --- Comparación sobre la sub-instancia demo (4 partidos) ---
# Usamos la misma jornada reducida que en la fuerza bruta para poder comparar
# ambos algoritmos sobre exactamente el mismo problema.
# Adaptamos demo al formato (loc, vis, tipo) que usa fo en esta celda,
# ya que la fuerza bruta de cell-14 usaba tuplas de 2 elementos y aquí usamos tuplas de 3.
demo_ext = [(None, None, p) for p in demo]
demo_cats = [p[2] for p in demo_ext]

print("Comparación sobre la jornada demo (4 partidos):")

sol_fb, fo_fb, _ = fuerza_bruta(demo_ext)
print(f"  Fuerza bruta:     {fo_fb:.4f} M  (óptimo garantizado)")

random.seed(42)
sol_bt, fo_bt_demo = busqueda_tabu(demo_cats, max_iter=500, tamanio_tabu=15)
print(f"  Búsqueda Tabú:    {fo_bt_demo:.4f} M")
print(f"  Diferencia:       {abs(fo_bt_demo - fo_fb):.4f} M ({abs((fo_bt_demo/fo_fb)-1)*100:.2f}% del óptimo)")

# --- Aplicación a la jornada completa (10 partidos) ---
# Para la jornada completa la fuerza bruta no es viable, así que ejecutamos
# solo la Búsqueda Tabú y mostramos el resultado.
print("\nAplicación a la jornada completa (10 partidos):")
print("Configuración de la Jornada:")
for i, (loc, vis, tipo) in enumerate(jornada):
    print(f" {i+1}. {loc} vs {vis} [{tipo[0]}-{tipo[1]}]")

t0 = time.time()
asig_bt, fo_bt = busqueda_tabu(cats, max_iter=500, tamanio_tabu=15)
t1 = time.time()

print(f"\nResultado Búsqueda Tabú: {fo_bt:.4f} M (Tiempo: {t1-t0:.2f}s)")

print("\nAsignación Detallada:")
cnt = Counter(asig_bt)
for i, h in enumerate(asig_bt):
    loc, vis, tipo = jornada[i]
    # Desglosamos el cálculo para verificar los coeficientes aplicados
    base = aud_base[tipo]
    ch = franjas[h][1]
    cc = penalizacion(cnt[h])
    aud = base * ch * cc
    print(f" {i+1}. {loc}-{vis}: {franjas[h][0]} | Base: {base} * Ch: {ch} * Cc: {cc} = {aud:.4f}M")

print(f"\nAUDIENCIA TOTAL FINAL: {fo_bt:.4f} M")

print("\nOcupación de franjas:")
for h_idx, count in sorted(cnt.items()):
    print(f" - {franjas[h_idx][0]}: {count} partido(s)")
````

## Posibles mejoras y líneas de trabajo futuro

### Mejoras al algoritmo

La principal limitación de la Búsqueda Tabú tal como la hemos implementado es que depende bastante del punto de partida inicial y del tamaño de la lista tabú. Hay varias formas de mejorarlo:

- **Multi-start Tabú**: lanzar varias ejecuciones desde puntos de partida aleatorios distintos y quedarse con la mejor solución. Combina la diversificación del multi-start con la capacidad de escapar de óptimos locales de Tabú.
- **Intensificación y diversificación adaptativa**: ajustar dinámicamente el tamaño de la lista tabú según el progreso del algoritmo. Si llevamos muchas iteraciones sin mejorar, aumentamos el tamaño para forzar más exploración; si estamos mejorando, lo reducimos para explotar la zona prometedora.
- **Simulated Annealing**: en lugar de prohibir movimientos, acepta soluciones peores con cierta probabilidad que va bajando con el tiempo. Para problemas de scheduling como éste suele dar buenos resultados con una implementación más sencilla.
- **Algoritmos Genéticos**: trabajan con una población de soluciones y aplican cruce y mutación. Más costosos computacionalmente pero pueden explorar mejor espacios muy grandes.

### Variaciones del problema

- **Temporada completa**: si en vez de una jornada consideramos las 38, aparecen restricciones nuevas de equidad (que ningún equipo juegue siempre en viernes o lunes) y de variedad en los horarios. La lista tabú podría adaptarse para registrar también combinaciones partido-horario a lo largo de toda la temporada.
- **Derechos televisivos**: en la realidad ciertos partidos tienen que emitirse en franjas concretas por contratos con operadores. Eso añade restricciones duras que habría que incorporar como criterios adicionales de validez.
- **Escalar a más equipos o ligas**: con más partidos por jornada el espacio crece exponencialmente y la fuerza bruta se vuelve todavía más inviable. Ahí es donde las metaheurísticas como Tabú realmente muestran su valor.
